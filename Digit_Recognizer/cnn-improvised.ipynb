```python
# splitting the data into train_data, train_labels
train_data = train.iloc[:, 1:]
train_label = train.iloc[:, 0]
```

```python
# Doing the train test split on train_data and train_label, we don't want to touch the 
# test data until out model is trained.
# from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(train_data, train_label, test_size=0.1)
```

```python
# Normalize the pixel intensity values
train_data = train_data.astype('float32') / 255
test = test.astype('float32') / 255
```

```python
# Reshape the array to shape (num, 28, 28, 1)
train_data = train_data.to_numpy().reshape(train_data.shape[0], 28, 28, 1)
test = test.to_numpy().reshape(test.shape[0], 28, 28, 1)

# convert class vectors to binary class matrices
train_label = tf.keras.utils.to_categorical(train_label, 10)
# y_test = keras.utils.to_categorical(y_test, 10)
```

```python
# Making CNN
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten

model = Sequential([
    Conv2D(filters=32, kernel_size=3, activation='relu'),
    MaxPooling2D(pool_size=2),
    Conv2D(filters=64, kernel_size=3, activation='relu'),
    MaxPooling2D(pool_size=2),
    Flatten(),
    layers.Dropout(0.5),
    Dense(120, activation='sigmoid'),
    Dense(84, activation='sigmoid'),
    Dense(10, activation='softmax'),
])
```

```python
# Training Phase
batch_size = 128
epochs = 15

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
model.fit(
    x=train_data,
    y=train_label,
    batch_size=batch_size,
    epochs=epochs,
    verbose=2,
    validation_split=0.1,
)
```

```python
preds = model.predict(test, verbose=2)
```
